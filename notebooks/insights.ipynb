```python
import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

output_dir = '../data/outputs'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password='1234',
    database='prouni_db'
)

query = "SELECT * FROM cursos_prouni"
df = pd.read_sql(query, conn)
conn.close()

display(df.head())
display(df.info())
display(df.describe())


```

<div style="text-align: center;">
  <h1>Contagem de cursos por tipo de grau</h1>
</div>

```python
grau_counts = df['grau'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=grau_counts.index, y=grau_counts.values, palette="viridis")
plt.xlabel("Tipo de Grau")
plt.ylabel("Quantidade de Cursos")
plt.title("Distribuição dos Tipos de Grau")
plt.xticks(rotation=45)
plt.savefig(os.path.join(output_dir, 'distribuicao_grau.jpg'), bbox_inches = 'tight')
plt.show()
```

<div style="text-align: center;">
  <h1> Distribuição das mensalidades</h1>
</div>


```python
plt.figure(figsize=(10,6))
sns.histplot(df['mensalidade'], bins=30, kde=True, color='purple')
plt.xlabel("Mensalidade (R$)")
plt.ylabel("Frequência")
plt.title("Distribuição das Mensalidades")
plt.savefig(os.path.join(output_dir, 'distribuicao_mensalidades.jpg'), bbox_inches = 'tight')
plt.show()
```

<div style="text-align: center;">
  <h1> Análise notas de corte para os cursos mais frequentes</h1>
</div>


```python
top_cursos = df['curso_busca'].value_counts().index[:10]
df_top_cursos = df[df['curso_busca'].isin(top_cursos)]

plt.figure(figsize=(12,6))
sns.boxplot(x='curso_busca', y='nota_integral_ampla', data=df_top_cursos, palette='coolwarm')
plt.xticks(rotation=45)
plt.xlabel("Curso")
plt.ylabel("Nota de Corte (Ampla)")
plt.title("Notas de corte para os Cursos Mais Frequentes")
plt.savefig(os.path.join(output_dir, 'notas_corte_cursos_frequentes.jpg'), bbox_inches = 'tight')
plt.show()
```

```python
df_path_exit = "../data/processed/cursos_prouni_direto_mysql.csv"
df.to_csv(df_path_exit, index=False)
print(f"Arquivo salvo em: {df_path_exit}")
```

<div style="text-align: center;">
  <h1> Distribuição de Cursos por Cidade</h1>
</div>

```python
city_counts = df['cidade_busca'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=city_counts.index, y=city_counts.values, palette='viridis')
plt.xlabel("Cidade")
plt.ylabel("Número de Cursos")
plt.title("Top 10 Cidades Com Mais Cursos")
plt.xticks(rotation=45)
plt.savefig(os.path.join(output_dir, 'cursos_cidade.jpg'), bbox_inches = 'tight')
plt.show()
```

<div style="text-align: center;">
  <h1>Distribuição de Mensalidades por Estado (Região)</h1>
</div>

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='uf_busca', y='mensalidade', data=df, palette="Set3", inner=None)
sns.swarmplot(x='uf_busca', y='mensalidade', data=df, color='k', alpha=0.6)
plt.xlabel("Estado")
plt.ylabel("Mensalidade (R$)")
plt.title("Distribuição de Mensalidades por Estado")
plt.tight_layout()
plt.show()
```
